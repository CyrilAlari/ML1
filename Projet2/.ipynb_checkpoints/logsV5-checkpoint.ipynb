```python
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv('data/train.csv')
train2 = pd.read_csv('data/train_v2.csv')
test = pd.read_csv('data/sample_submission_v2.csv')

union = pd.merge(train, train2, on='msno', how='outer')
union = pd.merge(union, test, on='msno', how='outer')
```

```python
msno_list = pd.concat([train2.msno,train.msno,test.msno]).unique()
```

```python
months = [(2017,2),(2017,1),(2016,12),(2016,11),(2016,10),(2016,9),(2016,8)]
selected= pd.DataFrame(index= [(i,y,m) for i in union.msno for(y,m) in months], columns = ["num_25_sum","num_50_sum","num_75_sum","num_985_sum","num_100_sum","num_unq_sum","total_secs_sum"])
selected = selected.fillna(0)
```

```python
logs_iter = pd.read_csv('data/user_logs.csv', low_memory=False, iterator=True, chunksize=1000000, parse_dates=["date"], infer_datetime_format = True)
ref_y = 2017
ref_m = 2
i = 0

for logs in logs_iter:
    print("step", i)
    new_logs = logs.merge(union, on='msno', how='right')
    new_logs = new_logs.loc[new_logs["date"] >= pd.to_datetime('20160801', format='%Y%m%d', errors='ignore')]
    new_logs.sort_values(["date"])
    new_logs['month'] = new_logs['date'].apply(lambda val:val.month)
    new_logs['year'] = new_logs['date'].apply(lambda val:val.year)
    new_logs = new_logs.drop(["date","is_churn_x","is_churn_y","is_churn"],axis=1)
    
    regroup_log = new_logs.groupby(by=["msno","year","month"]).agg(["sum"])
    regroup_log.index = [(i,y,m) for i,y,m in regroup_log.index]

    for ind in regroup_log.columns.get_level_values(0):
        selected[ind+"_sum"] = selected[ind+"_sum"] + regroup_log[ind]["sum"]
    i+=1
```

```python
final = pd.DataFrame(index= [i for i in union.msno], columns = ["num_25_last","num_50_last","num_75_last","num_985_last","num_100_last","num_unq_last","total_secs_last","num_unq_dyn","total_secs_dyn"])
```

```python
for col in selected.columns:
    for i in union.msno:
        final.loc[i][col+"_last"] = selected.loc[(i,2017,2)][col]

for col in [num_unq_dyn,total_secs_dyn]:
    for i in union.msno:
        final.loc[i][col+"_dyn"] = sum(selected.loc[(i,y,m)][col]/((2017-y)*12+(2-m)) for y,m in months)
    
```
